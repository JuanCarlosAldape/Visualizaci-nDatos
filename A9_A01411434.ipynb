```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
```

```python
modelA = sm.OLS.from_formula('peso ~ cintura', data = df)
resultA = modelA.fit()
resultA.summary()
```

```python
plt.figure(figsize = (15,10))
sns.set(font_scale = 1.7)

reg_plot = sns.regplot (x = 'peso',
             y = 'cintura',
             data = df,
             marker = '+',
             color = 'green',
            scatter_kws = {'s':200}
            )
reg_plot.set_title('Regression plot of Peso vs vs Cintura')
reg_plot.text(42, 100, 'Peso = -8.3137 + 0.8316 * cintura',
              color = 'green',
              weight = 'semibold',
              size = 'small'
             )
```

```python
RESIA =resultA.resid
RESIA
```
